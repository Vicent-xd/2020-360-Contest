```python
import numpy as np
import pandas as pd
import re
from os import walk
import csv
```

```python
#d = r"F:/2020DataCon比赛/dataCon2020/加密流量分析/eta_1/demo/white"
white_dir = r"F:\2020DataCon比赛\Outcsv_white\Outcsv_white"
black_dir = r"F:\2020DataCon比赛\Outcsv_black\Outcsv_black"

def pre_zj(ab_dir):                               #输入存储csv的绝对路径，输出dataframe
    ip_feature_tmp = pd.DataFrame(columns=('ip',
                                       'certificates_length', 
                                       'certificates_Num', 
                                       'certificates_Chain', 
                                       'self_signature', 
                                       ))

    for f, _, i in walk(ab_dir):
        for index,j in enumerate(i):
            #print(f)
            #print(index)
            dir=f+'\\'+j
            #print(dir)
            raw_tmp=['','','','','']
            raw_tmp[0]=j.replace('.csv','')

            with open(dir,'r') as csvfile:
                reader = csv.reader(csvfile)
                column_cert_len = [row[15]for row in reader]
            #print(column_cert_len)
            while '' in column_cert_len:
                column_cert_len.remove('')
            cerlist = list(set(column_cert_len[1:]))            #certificate_length字段 将列表中的元素无重复的抽取出来，赋值给另一个列表
            #print ((cerlist))
            #print(len(cerlist))                  #certificate_Num
            if len(cerlist)!=0:
                raw_tmp[1]=max(cerlist)
            else:
                raw_tmp[1]=0
            raw_tmp[2]=len(cerlist)              

            with open(dir,'r') as csvfile:
                reader = csv.reader(csvfile)
                column_issuer = [row[17]for row in reader]
            #print(column_issuer)
            while '' in column_issuer:
                column_issuer.remove('')
            issuer_len = list(set(column_issuer[1:])) 

            with open(dir,'r') as csvfile:
                reader = csv.reader(csvfile)
                cer_string = [row[21]for row in reader]
            #print(cer_string)
            while '' in cer_string:
                cer_string.remove('')
            cer_str_len = list(set(cer_string[1:])) 
            raw_tmp[3] = cer_str_len
            #print(raw_tmp[3])
            with open(dir,'r') as csvfile:
                reader = csv.reader(csvfile)
                column_subject = [row[23]for row in reader]
            #print(column_subject)
            while '' in column_subject:
                column_subject.remove('')
            subject_len = list(set(column_subject[1:])) 

            tmp=0               #self_signature
            if len(issuer_len)!=len(subject_len):
                print('证书签发与接收数量不一致')
            else:
                for i in range(len(issuer_len)):
                    if (issuer_len[i]==1)&(subject_len[i]==1)&(',' not in issuer_len[i]):
                        tmp=1
            raw_tmp[4] = tmp

            ip_feature_tmp.loc[index] = raw_tmp
    return ip_feature_tmp
        
white_zj=pre_zj(white_dir)
white_zj
```

```python
#bdir=r"F:\2020DataCon比赛\Outcsv_black\Outcsv_black"
black_zj=pre_zj(black_dir)
black_zj
```

```python
dict_black=[]
for i in black_zj['certificates_Chain'].values:
    for k in i:
        dict_black+=k.split(',')
len(dict_black)
b_tmp=(set(dict_black))

dict_white=[]
for i in white_zj['certificates_Chain'].values:
    for k in i:
        dict_white+=k.split(',')
len(dict_white)
w_tmp=(set(dict_white))
dict_all=list(set(list(w_tmp)+list(b_tmp)))
dict_all
```

```python
def list_add(a,b):
    c = []
    for i in range(len(a)):
        c.append(a[i]+b[i])
    return c
def feature_engineering(raw_str:str,CHARSET):
    char_feature = [raw_str.count(char) for char in CHARSET]
    return char_feature

def one(list_num):
    list_num1 = []
    for i in list_num:
        if i > 0:
            list_num1.append(1)
        else:
            list_num1.append(0)
    return list_num1
```

```python
mat_black = []
mat_white = []

for i in black_zj['certificates_Chain'].values:
    tmp_char = " ".join(i)
    line_vel=one(feature_engineering(tmp_char,dict_all))
    mat_black.append(line_vel)
    
for i in white_zj['certificates_Chain'].values:
    tmp_char = " ".join(i)
    line_vel=one(feature_engineering(tmp_char,dict_all))
    mat_white.append(line_vel)

sum_black = [0 for x in range(len(dict_all))]
sum_white = [0 for x in range(len(dict_all))]

for i in mat_black:
    sum_black = list_add(sum_black,i)
sum_black

for i in mat_white:
    sum_white = list_add(sum_white,i)

rows = zip(dict_all,sum_black,sum_white)
with open('test.csv', "w") as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
