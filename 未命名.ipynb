```python
import pandas as pd
import os

startdir = '../Outcsv_white1/'
suffix = '.csv'
outputdir = os.path.join(startdir, 'all_white')

list_ip = []
# list_cipher_spec_len = []
# list_cipher_spec_num = []
# list_cipher_suites_len = []
# list_cipher_suite_num = []
list_record_version = []
list_handshake_version = []
list_x509af_extensions = []
i=1
for root, dirs, files in os.walk(startdir):
    # print(root,dirs,files)
    for name in files:
        if name=='.DS_Store':
            continue;
        IP = os.path.splitext(name)[0]
        list_ip.append(IP)
        print(IP)
        i=i+1
        print(i)
        raw_data = pd.read_csv(startdir+name)
        temp_data1 = raw_data.copy(deep=True)
        temp_data2 = raw_data.copy(deep=True)
        temp_data3 = raw_data.copy(deep=True)
        temp_data4 = raw_data.copy(deep=True)
        temp_data_record_version = raw_data.copy(deep=True)
        temp_data_handshake_version = raw_data.copy(deep=True)
        temp_data_x509af_extensions = raw_data.copy(deep=True)


        # temp_data_record_version
        temp_data_record_version.dropna(
            subset=['tls.record.version'], inplace=True)
        if temp_data_record_version.shape[0] == 0:
            list_record_version.append(-1)
        else:
            record_version_unique = temp_data_record_version['tls.record.version'].unique(
            )
            if len(record_version_unique) == 1:
                list_record_version.append(record_version_unique[0])
            elif len(record_version_unique) < 1:
                list_record_version.append(-1)
            else:
                 # list_cipher_spec_num.append(cipher_spec_num_unique)
                list_record_version_t = []
                for record_version_t in record_version_unique:
                    # print(type(suite))
                    # print(suite)
                    if type(record_version_t) == type('str'):
                        temp_list = record_version_t.split(',')
                    else:
                        temp_list = record_version_t.astype('str').split(',')
                    list_record_version_t = list_record_version_t+temp_list
                # print(list_spec)
                set_record_version_t = set(list_record_version_t)
                # print(set_spec)
                list_record_version.append(set_record_version_t)

        # temp_data_handshake_version
        temp_data_handshake_version.dropna(
            subset=['tls.handshake.version'], inplace=True)
        if temp_data_handshake_version.shape[0] == 0:
            list_handshake_version.append(-1)
        else:
            handshake_version_unique = temp_data_handshake_version['tls.handshake.version'].unique(
            )
            if len(handshake_version_unique) == 1:
                list_handshake_version.append(handshake_version_unique[0])
            elif len(handshake_version_unique) < 1:
                list_handshake_version.append(-1)
            else:
                 # list_cipher_spec_num.append(cipher_spec_num_unique)
                list_handshake_version_t = []
                for handshake_version_t in handshake_version_unique:
                    # print(type(suite))
                    # print(suite)
                    if type(handshake_version_t) == type('str'):
                        temp_list = handshake_version_t.split(',')
                    else:
                        temp_list = handshake_version_t.astype('str').split(',')
                    list_handshake_version_t = list_handshake_version_t+temp_list
                # print(list_spec)
                set_handshake_version_t = set(list_handshake_version_t)
                # print(set_spec)
                list_handshake_version.append(set_handshake_version_t)

        # temp_data_x509af_extensions
        temp_data_x509af_extensions.dropna(
            subset=['x509af.extensions'], inplace=True)
        if temp_data_x509af_extensions.shape[0] == 0:
            list_x509af_extensions.append(-1)
        else:
            x509af_extensions_unique = temp_data_x509af_extensions['x509af.extensions'].unique(
            )
            list_x509af_extensions.append(x509af_extensions_unique)

new_data1 = pd.DataFrame({'IP': list_ip,'record_version': list_record_version,'handshake_version': list_handshake_version,'x509af_extensions': list_x509af_extensions})
new_data1.to_csv('../Outcsv_black/whitev1.csv', index=False)
print(new_data1)
```
